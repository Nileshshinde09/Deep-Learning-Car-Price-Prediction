```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("car_data.csv")
```

```python
dataset.head()
```

<h4>Encoding</h4> 

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset["Fuel_Type"] = le.fit_transform(dataset["Fuel_Type"])
dataset["Seller_Type"] = le.fit_transform(dataset["Seller_Type"])
dataset["Transmission"] = le.fit_transform(dataset["Transmission"])
```

<h4>Droping Car_Name Column</h4>

```python
dataset = dataset.drop(columns ="Car_Name",axis = 1)
```

```python
dataset.isnull().sum()
```

```python
x = dataset.drop(columns ="Owner",axis = 1)
x.head()
```

```python
y = dataset["Owner"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 90)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train.iloc[:,4:].head().values
x_train[["Fuel_Type","Seller_Type","Transmission"]]
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model= keras.Sequential([
    keras.layers.Dense(7,input_dim = 7,activation = 'relu'),
    keras.layers.Dense(14,activation = 'relu'),
    keras.layers.Dense(14,activation = 'relu'),
    keras.layers.Dense(7,activation = 'relu'),
    keras.layers.Dense(1,activation = 'sigmoid')
])
model.compile(loss = 'binary_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
model.fit(x_train,y_train,epochs = 200,batch_size = 2)
```

accuracy  = 95%

```python
from tensorflow.keras.models import save_model
save_model(model, "model.h5")
```
